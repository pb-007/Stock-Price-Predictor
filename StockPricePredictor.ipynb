```python
import numpy as np
```

```python
df1 = df.reset_index()['Close']
```

```python
df1.shape
```

```python
from sklearn.preprocessing import MinMaxScaler
scaled_data = MinMaxScaler(feature_range = (0,1))
df1 = scaled_data.fit_transform(np.array(df1).reshape(-1,1))
```

```python
df1
```

```python
training_size = int(len(df1)*0.70)
test_size = len(df1)-training_size
train_data, test_data = df1[0:training_size,:], df1[training_size: len(df1),:1]
```

```python
train_data.shape, test_data.shape
```

```python
print(train_data)
```

```python
import numpy
def create_dataset(dataset, time_step = 1):
  dataX,dataY = [],[]
  for i in range(len(dataset)- time_step-1):
    a = dataset[i:(i+time_step),0]
    dataX.append(a)
    dataY.append(dataset[i+time_step, 0])
  return numpy.array(dataX), numpy.array(dataY)  
```

```python
time_step = 200
X_train, Y_train = create_dataset(train_data, time_step)
X_test, Y_test = create_dataset(test_data, time_step)
```

```python
print(X_train)
```

```python
print(Y_train)
```

```python
X_train = X_train.reshape(X_train.shape[0],X_train.shape[1], 1)
X_test = X_test.reshape(X_test.shape[0], X_test.shape[1], 1)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import LSTM
```

```python
model = Sequential()
model.add(LSTM(50,return_sequences = True,input_shape=(200,1)) )
model.add(LSTM(50,return_sequences = True))
model.add(LSTM(50))
model.add(Dense(1))
model.compile(loss = 'mean_squared_error', optimizer = 'adam')
```

```python
model.summary()
```

```python
model.fit(X_train,Y_train,validation_data=(X_test,Y_test), epochs = 200, batch_size = 69, verbose= 1)
```

```python
train_predict = model.predict(X_train)
test_predict = model.predict(X_test)
```

```python
#transform into its original form for calculating the root mean square error
train_predict = scaled_data.inverse_transform(train_predict)
test_predict = scaled_data.inverse_transform(test_predict)
```

```python
import math
from sklearn.metrics import mean_squared_error
math.sqrt(mean_squared_error(Y_train,train_predict))
```

```python
math.sqrt(mean_squared_error(Y_test,test_predict))
```

```python
import matplotlib.pyplot as plt
look_back = 200
trainpredictplot = numpy.empty_like(df1)
trainpredictplot[:,:]= np.nan
trainpredictplot[look_back:len(train_predict)+look_back,:]=train_predict

testpredictplot = numpy.empty_like(df1)
testpredictplot[:,:] = numpy.nan
testpredictplot[len(train_predict)+(look_back*2)+ 1:len(df1)-1,:] = test_predict
plt.plot(scaled_data.inverse_transform(df1))
plt.plot(trainpredictplot)
plt.plot(testpredictplot)
plt.figure(figsize=(30,20))
plt.show
```
